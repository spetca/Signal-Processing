```python
def upsample(seq,length,by):
    cnt = 0
    seqcnt = 0
    x = np.zeros(length)
    for n in range(0,length):
        if(cnt % by == 0):
            x[n] = seq[seqcnt]
            cnt = cnt +1
            seqcnt = seqcnt + 1
        else:
            x[n] = 0
            cnt + 1
    return x
```

A communication channel is formed by a cascade of four processes as shown below. The
matched filter operates with the shaping filter to simultaneously maximize signal to noise ratio (SNR)
and minimize inter symbol interference (ISI). The channel induced distortion introduces ISI and the
equalizer serves to minimize its effect.

<img src="files/2.png">

The input is 1000 samples of a complex QPSK random data sequence, the shaping filter is a square root Nyquist filter with excess bandwidth  = 0.25 operating at 4-samples per symbol, the channel impulse response is {1 0 0 0 0 0.2 0 0 j*0.1}, and the noise is complex AWGN with $\sigma^{2} =0.01$.
The equalizer is a 40 tap FIR filter operating at 4-samples per symbol and is updated at symbol rate

```python
x0=(np.floor(2*random.random(1000,))-0.5)/0.5+1j*(np.floor(2*random.random(1000,))-0.5)/0.5
sigma_sq = .01/2
noiseR  =  (sqrt(sigma_sq)*random.standard_normal((4000,)))
noiseI = sqrt(sigma_sq)*random.standard_normal((4000,))
noise = noiseR + 1j*noiseI
chan = np.array([1, 0, 0, 0, 0, 0.2, 0, 0, 1j*0.1])

octave.addpath('files/')
h0 = octave.srrc(4, .25,10)

h0 = np.asarray(h0).flatten()
h1 = h0/np.max(h0);

x1= np.vstack((x0, np.zeros(shape=(3,1000)))).flatten(1)

x2= signal.lfilter(h1,1,x1)
xc = signal.lfilter(chan, 1,x2) + noise;
xmf = signal.lfilter(h1,1,xc)/(np.asmatrix(h1)*np.asmatrix(h1).transpose());
```
