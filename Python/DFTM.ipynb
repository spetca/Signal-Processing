<H1>Dual Frequency Tone Modulation Example</h1>

Create 8 tones associated with a phone keypad. These are different magnitudes so we can track them as we process the data below. 
<img src="http://ptolemy.eecs.berkeley.edu/eecs20/week2/keypad.gif" alt="">

```python
from numpy import *
from scipy import *
from scipy import signal
from scipy.signal  import butter, lfilter, tf2ss
import scipy.linalg
import matplotlib.pyplot as plt

Fs9k=9000 #sampling frequency

#amplitudes and frequencies for test tone
a1 = 1.0
f1 = 697

a2 = .5
f2 = 770

a3 = .25
f3 = 852

a4 = .125
f4 = 941

a5 = 1.0
f5 = 1209

a6 = .5
f6 = 1336

a7 = .25
f7 = 1477

a8 = .125
f8 = 1633

#20000 point test tones
tone1_9k = a1*sin(2*pi*f1/Fs9k*arange(1., 20001.))
tone2_9k = a2*sin(2*pi*f2/Fs9k*arange(1., 20001.))
tone3_9k = a3*sin(2*pi*f3/Fs9k*arange(1., 20001.))
tone4_9k = a4*sin(2*pi*f4/Fs9k*arange(1., 20001.))
tone5_9k = a5*sin(2*pi*f5/Fs9k*arange(1., 20001.))
tone6_9k = a6*sin(2*pi*f6/Fs9k*arange(1., 20001.))
tone7_9k = a7*sin(2*pi*f7/Fs9k*arange(1., 20001.))
tone8_9k = a7*sin(2*pi*f8/Fs9k*arange(1., 20001.))

#test signal
test_signal = tone1_9k+tone2_9k+tone3_9k+tone4_9k+tone5_9k+tone6_9k+tone7_9k+tone8_9k

plt.plot(arange(1., 20001.),test_signal)
plt.title('Test Signal')
```

<h2>Design a Elliptical 9th order lowpass filter with a sampling rate of 9kHz for preprocessing signals</h2>

```python
fpass = 1621;
fstop = 2670;

[b, a] = scipy.signal.ellip(9, .1, 60, fpass/(Fs9k/2))
x0 = np.zeros(2048)
x0[0] = 1

#impulse response of the filter
h1 = signal.lfilter(b,a,x0)

plt.plot(arange(1,2049),h1)
plt.title('Impulse response 9th order elliptical filter')
pylab.xlim(0, 120)
pylab.ylim(-.5, .5)
```

```python
#fft of impulse response is the filter response
fft_h1 = 20*np.log10(abs(scipy.fftpack.fft(h1, 2048)))

plt.plot(arange(1.,2049.)*Fs9k/2048, fft_h1)
plt.title('lowpass filter')
#plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([0, 4500, -120, 3])
plt.grid(True)
plt.show()
```

```python
ww = numpy.kaiser(2048, 10)
ww = ww/sum(ww)
ww = ww.conj().T
```

<h2>Windowed Spectra of test signals:</h2>

```python
FFT_testsignal = 20*np.log10(abs(scipy.fftpack.fft(test_signal[1:2049] * ww, 2048)))

subplot(2,1,1)
plt.plot(arange(1.,2049.)*Fs9k/2048, FFT_testsignal)
plt.title('Test signal Spectra')
#plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([0, 4500, -60, 50])
plt.grid(True)
plt.show()

subplot(2,1,2)
plt.plot(arange(1.,2049.)*Fs9k/2048, FFT_testsignal)
plt.title('Zoom Test signal Spectra')
#plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([0, 2000, -60, 50])
plt.grid(True)
plt.show()
```

<h2>Filter signals with 9kHz low pass filter then decimate 2-to-1:</h2>

```python
x1 = lfilter(b,a,test_signal)
x2 = scipy.signal.decimate(x1, 2)

FFT_x1 = 20*np.log10(abs(scipy.fftpack.fft(x1[1:2049]*ww, 2048)))
FFT_x2 = 20*np.log10(abs(scipy.fftpack.fft(x2[1:2049]*ww, 2048)))

subplot(2,1,1)
plt.plot(arange(1.,2049.)*Fs9k/2048, FFT_x1)
plt.title('x1 - filtered data')
#plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([0, 4500, -60, 10])
plt.grid(True)
plt.show()

subplot(2,1,2)
plt.plot(arange(1.,2049.)*Fs9k/2048, FFT_x2)
plt.title('x2 - 2-to-1 downsampled data')
#plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([0, 4500, -60, 30])
plt.grid(True)
plt.show()
```

<h2>Design low pass and high pass filters to split the row and column frequencies:</h2>

```python
fs4_5k=4500;

tone1_4_5k = a1*sin(2*pi*f1/fs4_5k*arange(1., 20001.));
tone2_4_5k = a2*sin(2*pi*f2/fs4_5k*arange(1., 20001.));
tone3_4_5k = a3*sin(2*pi*f3/fs4_5k*arange(1., 20001.));
tone4_4_5k = a4*sin(2*pi*f4/fs4_5k*arange(1., 20001.));
tone5_4_5k = a5*sin(2*pi*f5/fs4_5k*arange(1., 20001.));
tone6_4_5k = a6*sin(2*pi*f6/fs4_5k*arange(1., 20001.));
tone7_4_5k = a7*sin(2*pi*f7/fs4_5k*arange(1., 20001.));
tone8_4_5k = a8*sin(2*pi*f8/fs4_5k*arange(1., 20001.));

testsig_4_5k = tone1_4_5k+tone2_4_5k+tone3_4_5k+tone4_4_5k+tone5_4_5k+tone6_4_5k+tone7_4_5k+tone8_4_5k;

fpasslow = 950;
fstoplow = 1200;

#lowpass 2nd stage
[b_low, a_low] = scipy.signal.ellip(8, .1 , 60, 932.99/(fs4_5k/2));
h_low = lfilter(b_low,a_low,x0);
FFT_hlow = 20*np.log10(abs(scipy.fftpack.fft(h_low, 2048)))

#high pass 2nd stage
[b_high, a_high] = scipy.signal.ellip(8, .1 , 60, 1217.4/(fs4_5k/2), 'high');
h_high = lfilter(b_high,a_high,x0);
FFT_hhigh= 20*np.log10(abs(scipy.fftpack.fft(h_high, 2048)))

plt.plot(arange(1.,2049.)*Fs45k/2048, FFT_hlow)
plt.title('2nd stage : lowpass filter')
#plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([0, 2250, -120, 3])
plt.grid(True)
plt.show()

plt.plot(arange(1.,2049.)*Fs45k/2048, FFT_hhigh)
plt.title('2nd stage : lowpass filter')
#plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([0, 2250, -120, 3])
plt.grid(True)
plt.show()
```

<h2>Filter signals with low and high pass filter:</h2>

```python
low_filter = lfilter(b_low,a_low,x2);
high_filter = lfilter(b_high, a_high, x2);


FFT_low_filter = 20*np.log10(abs(scipy.fftpack.fft(low_filter[1:2049]*ww,2048)))
FFT_high_filter = 20*np.log10(abs(scipy.fftpack.fft(high_filter[1:2049]*ww, 2048)))



plt.plot(arange(1.,2049.)*Fs45k/2048, FFT_low_filter)
plt.plot(arange(1.,2049.)*Fs45k/2048, FFT_hlow)
plt.title('Spectrum of filtered data x2 through Low Pass Filter')
plt.axis([0, 2250, -120, 3])
plt.grid(True)
plt.show()


plt.plot(arange(1.,2049.)*Fs45k/2048, FFT_high_filter)
plt.plot(arange(1.,2049.)*Fs45k/2048, FFT_hhigh)
plt.title('Spectrum of filtered data x2 through High Pass Filter')
plt.axis([0, 2250, -120, 3])
plt.grid(True)
plt.show()

```

<h2>Decimate 4-to-1 , sampling rate now 1125Hz</h2>

```python
Fs1125 = 1125
ds4to1_lpf = scipy.signal.decimate(low_filter,4);
ds4to1_hpf = scipy.signal.decimate(high_filter,4);

FFT_ds4to1_lpf = 20*np.log10(abs(scipy.fftpack.fft(ds4to1_lpf[1:2049]*ww,2048)))
FFT_ds4to1_hpf = 20*np.log10(abs(scipy.fftpack.fft(ds4to1_hpf[1:2049]*ww,2048)))

plt.plot(arange(1.,2049.)*Fs1125/2048, FFT_ds4to1_lpf)
plt.title(' 4-to-1 downsampled and Through Low Pass filter')
plt.axis([0, Fs1125/2, -200, 3])
plt.grid(True)
plt.show()

plt.plot(arange(1.,2049.)*Fs1125/2048, FFT_ds4to1_hpf)
plt.title('4-to-1 downsampled and Through High Pass Filter ')
plt.axis([0, Fs1125/2, -250, 100])
plt.grid(True)
plt.show()
```

<h2>Design bank of 8 bandpass filter to seperate tones</h2>

```python
f1_pos = 186.6
f2_pos = 274
f3_pos = 358
f4_pos = 430

wn1 = np.array([(f1_pos-8)/(Fs1125/2),  (f1_pos+8)/(Fs1125/2)])
wn2 = np.array([(f2_pos-8)/(Fs1125/2),  (f2_pos+8)/(Fs1125/2)])
wn3 = np.array([(f3_pos-8)/(Fs1125/2),  (f3_pos+8)/(Fs1125/2)])
wn4 = np.array([(f4_pos-8)/(Fs1125/2),  (f4_pos+8)/(Fs1125/2)])

[b_bp_f1 ,a_bp_f1] = scipy.signal.ellip(8, .1, 60, wn1,btype='bandpass')
h_bp1 = lfilter(b_bp_f1, a_bp_f1, x0)
FFT_bp1 = 20*np.log10(abs(scipy.fftpack.fft(h_bp1)))


[b_bp_f2, a_bp_f2] = scipy.signal.ellip(8, .1,60, wn2,btype='bandpass')
h_bp2 = lfilter(b_bp_f2, a_bp_f2, x0)
FFT_bp2 = 20*np.log10(abs(scipy.fftpack.fft(h_bp2)))

[b_bp_f3 ,a_bp_f3] = scipy.signal.ellip(8, .1,60, wn3,btype='bandpass')
h_bp3 = lfilter(b_bp_f3, a_bp_f3, x0)
FFT_bp3 = 20*np.log10(abs(scipy.fftpack.fft(h_bp3)))

[b_bp_f4, a_bp_f4] = scipy.signal.ellip(8, .1,60, wn4,btype='bandpass')
h_bp4 = lfilter(b_bp_f4, a_bp_f4, x0)
FFT_bp4 = 20*np.log10(abs(scipy.fftpack.fft(h_bp4)))

plt.plot(arange(1,2049)*(1125/2048),FFT_ds4to1_lpf)
plt.plot(arange(1,2049)*(1125/2048), FFT_bp1, color='y')
plt.plot(arange(1,2049)*(1125/2048), FFT_bp2, color='r')
plt.plot(arange(1,2049)*(1125/2048), FFT_bp3, color='g')
plt.plot(arange(1,2049)*(1125/2048), FFT_bp4, color='m')
plt.axis([0, Fs1125/2, -150, 3])
plt.grid(True)
plt.show()
```
